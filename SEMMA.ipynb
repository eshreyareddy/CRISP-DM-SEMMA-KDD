```python
import pandas as pd

# Load the marketing_campaign dataset
marketing_campaign_df = pd.read_csv("marketing_campaign.csv")

# Display the first few rows of the dataset to get an overview
marketing_campaign_df.head()
```

```python
# Load the marketing_campaign dataset using tab as the delimiter
marketing_campaign_df = pd.read_csv("marketing_campaign.csv", delimiter="\t")

# Display the first few rows of the dataset to get an overview
marketing_campaign_df.head()
```

```python
# Display summary statistics for the dataset
marketing_campaign_summary = marketing_campaign_df.describe(include='all')

marketing_campaign_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the figure size
plt.figure(figsize=(15, 12))

# Distribution of Income
plt.subplot(2, 2, 1)
sns.histplot(marketing_campaign_df['Income'], bins=30, kde=True)
plt.title('Distribution of Income')

# Distribution of Education categories
plt.subplot(2, 2, 2)
sns.countplot(data=marketing_campaign_df, y='Education', order=marketing_campaign_df['Education'].value_counts().index)
plt.title('Distribution of Education Categories')

# Distribution of Marital_Status categories
plt.subplot(2, 2, 3)
sns.countplot(data=marketing_campaign_df, y='Marital_Status', order=marketing_campaign_df['Marital_Status'].value_counts().index)
plt.title('Distribution of Marital Status Categories')

# Distribution of Response
plt.subplot(2, 2, 4)
sns.countplot(data=marketing_campaign_df, x='Response')
plt.title('Distribution of Response to Campaign Offer')

plt.tight_layout()
plt.show()
```

```python
# Identify columns with missing values
missing_values = marketing_campaign_df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

missing_values
```

```python
# Fill missing values in the 'Income' column with the median value of the column
median_income = marketing_campaign_df['Income'].median()
marketing_campaign_df['Income'].fillna(median_income, inplace=True)

# Check if there are any more missing values
remaining_missing_values = marketing_campaign_df.isnull().sum().sum()

remaining_missing_values
```

```python
# One-hot encode the 'Education' and 'Marital_Status' columns
marketing_campaign_df_encoded = pd.get_dummies(marketing_campaign_df, columns=['Education', 'Marital_Status'], drop_first=True)

# Display the first few rows of the encoded dataset
marketing_campaign_df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split

# Define the feature set and target variable
X = marketing_campaign_df_encoded.drop(columns=['ID', 'Dt_Customer', 'Response'])
y = marketing_campaign_df_encoded['Response']

# Split the data into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize and train a logistic regression model
logreg_model = LogisticRegression(max_iter=1000, random_state=42)
logreg_model.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg_model.predict(X_test)

# Calculate accuracy on the test set
accuracy = accuracy_score(y_test, y_pred)

accuracy
```

```python
# Train and predict using the best parameters
best_model = LogisticRegression(C=10, penalty='l1', solver='liblinear', max_iter=1000, random_state=42)
best_model.fit(X_train, y_train)
y_pred_best = best_model.predict(X_test)

# Calculate accuracy on the test set for the best model
accuracy_best = accuracy_score(y_test, y_pred_best)

accuracy_best
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_best)

# Generate classification report
class_report = classification_report(y_test, y_pred_best)

conf_matrix, class_report
```
